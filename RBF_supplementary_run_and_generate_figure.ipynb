```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib

matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'font.size' : 8,
    'text.usetex': True,
    'pgf.rcfonts': False,
})
#matplotlib.rcParams.update(matplotlib.rcParamsDefault)
#matplotlib.rcParams.update({'font.size': 12})
#matplotlib.rcdefaults()
```

```python

fig,ax=plt.subplots(4,4,figsize=(40,60))

ms=4
lw=0.5
lwgt=1
mew=0.7

ylog=True
ls='--'
eb_alpha=0.5

alphaGT=0.5
cGT='b'
#fig,ax=plt.subplots(2,2,figsize=(6,6))
#ax.plot(Ps,gims,c='k',label='theory')




#casen=1

for ci in np.arange(4):

    # Load the .npz file
    data = np.load('RBF_plot_data_sup_BV_{}.npz'.format(ci+1))

    # Extract and assign each array to its original variable name
    kmax = data['kmax']
    seed = data['seed']
    sigmax = data['sigmax']

    P = data['P']
    Q = data['Q']
    sigma = data['sigma']
    d = data['d']

    Mss = data['Mss']
    Mns = data['Mns']
    Mk0s = data['Mk0s']
    Mk1s = data['Mk1s']
    Mts = data['Mts']

    # Optional: Close the .npz file
    data.close()





    xoi=np.arange(kmax-1)+2
    qoi=Mss
    colorv='r'
    zo=10
    yval=np.nanmedian(qoi,axis=1)
    lower_error=np.nanquantile(qoi,axis=1,q=0.25)
    upper_error=np.nanquantile(qoi,axis=1,q=0.75)
    asymmetric_error = [yval-lower_error, upper_error-yval]
    ax[ci][0].errorbar(xoi, yval, yerr=asymmetric_error,c=colorv,marker='',ls='',alpha=eb_alpha,lw=lw,zorder=zo)#, fmt='o')
    yval=np.nanmean(qoi,axis=1)
    #ax[0][0].plot(xoi, yval,c=colorv,marker='x',ls=ls,alpha=0.75,ms=10,lw=lw,label=r'Ours',zorder=zo*100)#, fmt='o')
    ax[ci][0].plot(xoi, yval,c=colorv,marker='x',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'Ours',zorder=zo*100,markeredgewidth=mew)#, fmt='o')


    qoi=Mns
    colorv='k'
    zo=5
    yval=np.nanmedian(qoi,axis=1)
    lower_error=np.nanquantile(qoi,axis=1,q=0.25)
    upper_error=np.nanquantile(qoi,axis=1,q=0.75)
    asymmetric_error = [yval-lower_error, upper_error-yval]
    ax[ci][0].errorbar(xoi, yval, yerr=asymmetric_error,c=colorv,marker='',ls='',alpha=eb_alpha,lw=lw,zorder=zo)#, fmt='o')
    yval=np.nanmean(qoi,axis=1)
    #ax[0][0].plot(xoi, yval,c=colorv,marker='x',ls=ls,alpha=0.75,ms=10,lw=lw,label=r'Naive',zorder=zo*100)#, fmt='o')
    ax[ci][0].plot(xoi, yval,c=colorv,marker='o',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'Naive',zorder=zo*100,markeredgewidth=mew)#, fmt='o')


    qoi=Mk0s
    colorv='g'
    zo=1
    yval=np.nanmedian(qoi,axis=1)
    lower_error=np.nanquantile(qoi,axis=1,q=0.25)
    upper_error=np.nanquantile(qoi,axis=1,q=0.75)
    asymmetric_error = [yval-lower_error, upper_error-yval]
    ax[ci][0].errorbar(xoi, yval, yerr=asymmetric_error,c=colorv,marker='',ls='',alpha=eb_alpha,lw=lw,zorder=zo)#, fmt='o')
    yval=np.nanmean(qoi,axis=1)
    #ax[0][0].plot(xoi, yval,c=colorv,marker='+',ls=ls,alpha=0.75,ms=10,lw=lw,label=r'KV row',zorder=zo*100)#, fmt='o')
    ax[ci][0].plot(xoi, yval,c=colorv,marker='>',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'KV row',zorder=zo*100,markeredgewidth=mew)#, fmt='o')



    qoi=Mk1s
    colorv='g'
    zo=1
    yval=np.nanmedian(qoi,axis=1)
    lower_error=np.nanquantile(qoi,axis=1,q=0.25)
    upper_error=np.nanquantile(qoi,axis=1,q=0.75)
    asymmetric_error = [yval-lower_error, upper_error-yval]
    ax[ci][0].errorbar(xoi, yval, yerr=asymmetric_error,c=colorv,marker='',ls='',alpha=eb_alpha,lw=lw,zorder=zo)#, fmt='o')
    yval=np.nanmean(qoi,axis=1)
    #ax[0][0].plot(xoi, yval,c=colorv,marker='x',ls=ls,alpha=0.75,ms=10,lw=lw,label=r'KV col',zorder=zo*100)#, fmt='o')
    ax[ci][0].plot(xoi, yval,c=colorv,marker='<',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'KV col',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    zo=-1
    ax[ci][0].plot(xoi,np.array(Mts),c=cGT,lw=lwgt,alpha=alphaGT,label='GT',zorder=zo)

    
    if ci==0:
        fig.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=8)


    ###############################



    gt=np.array(Mts)

    #fig,ax=plt.subplots(1,3,figsize=(12,4))
    #ax.plot(Ps,gims,c='k',label='theory')

    xoi=np.arange(kmax-1)+2
    qoi=Mss
    yval=np.nanmean(np.square(gt[:,None]-qoi),axis=1)
    colorv='r'
    zo=10
    ax[ci][1].plot(xoi, yval,c=colorv,marker='x',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'Ours',zorder=zo*100,markeredgewidth=mew)#, fmt='o')


    xoi=np.arange(kmax-1)+2
    qoi=Mns
    yval=np.nanmean(np.square(gt[:,None]-qoi),axis=1)
    colorv='k'
    zo=5
    ax[ci][1].plot(xoi, yval,c=colorv,marker='o',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'Naive',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    xoi=np.arange(kmax-1)+2
    qoi=Mk0s
    yval=np.nanmean(np.square(gt[:,None]-qoi),axis=1)
    colorv='g'
    zo=1
    ax[ci][1].plot(xoi, yval,c=colorv,marker='>',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'KV row',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    xoi=np.arange(kmax-1)+2
    qoi=Mk1s
    yval=np.nanmean(np.square(gt[:,None]-qoi),axis=1)
    colorv='g'
    zo=1
    ax[ci][1].plot(xoi, yval,c=colorv,marker='<',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'KV col',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    #ax.set_xscale('log')
    #ax[0][1].set_xlabel('n')
    #ax[1].legend(fontsize=11)
    #ax.set_title('N={}, Q={}'.format(N,Q),fontsize=11)




    xoi=np.arange(kmax-1)+2
    qoi=Mss
    yval=np.abs(gt-np.nanmean(qoi,axis=1))
    colorv='r'
    zo=10
    ax[ci][2].plot(xoi, yval,c=colorv,marker='x',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'Ours',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    xoi=np.arange(kmax-1)+2
    qoi=Mns
    yval=np.abs(gt-np.nanmean(qoi,axis=1))
    colorv='k'
    zo=5
    ax[ci][2].plot(xoi, yval,c=colorv,marker='o',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'Naive',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    xoi=np.arange(kmax-1)+2
    qoi=Mk0s
    yval=np.abs(gt-np.nanmean(qoi,axis=1))
    colorv='g'
    zo=1
    ax[ci][2].plot(xoi, yval,c=colorv,marker='>',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'KV row',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    xoi=np.arange(kmax-1)+2
    qoi=Mk1s
    yval=np.abs(gt-np.nanmean(qoi,axis=1))
    colorv='g'
    zo=1
    ax[ci][2].plot(xoi, yval,c=colorv,marker='<',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'KV col',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    #ax.set_xscale('log')
    #ax[0][2].set_xlabel('n')
    #ax[1].legend(fontsize=11)
    #ax.set_title('N={}, Q={}'.format(N,Q),fontsize=11)



    #ax.plot(Ps,gims,c='k',label='theory')

    xoi=np.arange(kmax-1)+2
    qoi=Mss
    yval=np.square(np.nanstd(qoi,axis=1))
    colorv='r'
    zo=10
    ax[ci][3].plot(xoi, yval,c=colorv,marker='x',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'Ours',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    xoi=np.arange(kmax-1)+2
    qoi=Mns
    yval=np.square(np.nanstd(qoi,axis=1))
    colorv='k'
    zo=5
    ax[ci][3].plot(xoi, yval,c=colorv,marker='o',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'Naive',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    xoi=np.arange(kmax-1)+2
    qoi=Mk0s
    yval=np.square(np.nanstd(qoi,axis=1))
    colorv='g'
    zo=1
    ax[ci][3].plot(xoi, yval,c=colorv,marker='>',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'KV row',zorder=zo*100,markeredgewidth=mew)#, fmt='o')

    xoi=np.arange(kmax-1)+2
    qoi=Mk1s
    yval=np.square(np.nanstd(qoi,axis=1))
    colorv='g'
    zo=1
    ax[ci][3].plot(xoi, yval,c=colorv,marker='<',fillstyle='none',ls='--',alpha=0.75,ms=ms,lw=lw,label=r'KV col',zorder=zo*100,markeredgewidth=mew)#, fmt='o')



    #ax.set_title('N={}, Q={}'.format(N,Q),fontsize=11)

    ax[ci][0].set_title(r'$\hat{m}(n)$')
    ax[ci][1].set_title('MSE')
    ax[ci][2].set_title('Bias')
    ax[ci][3].set_title('Variance')


from matplotlib.ticker import MaxNLocator
for i in range(np.shape(ax)[0]):
    for j in range(np.shape(ax)[1]):
        ax[i][j].xaxis.set_major_locator(MaxNLocator(integer=True))
        ax[i][j].set_yscale('log')
        ax[i][j].set_xlabel('n')
#ax[1][0].xaxis.set_major_locator(MaxNLocator(integer=True))
#ax[2][0].xaxis.set_major_locator(MaxNLocator(integer=True))
#ax[3][0].xaxis.set_major_locator(MaxNLocator(integer=True))

#ax[0][0].legend(fontsize=10).set_zorder(-100000)


fig.set_size_inches(w=6.75133, h=7)

fig.tight_layout()


bbox = ax[0][0].get_tightbbox(fig.canvas.get_renderer())
fig.text(bbox.x0, bbox.y1, 'a.', fontsize=15, fontweight="bold", va="top", ha="left",
             transform=None)
    
bbox = ax[1][0].get_tightbbox(fig.canvas.get_renderer())
fig.text(bbox.x0, bbox.y1, 'b.', fontsize=15, fontweight="bold", va="top", ha="left",
             transform=None)


bbox = ax[2][0].get_tightbbox(fig.canvas.get_renderer())
fig.text(bbox.x0, bbox.y1, 'c.', fontsize=15, fontweight="bold", va="top", ha="left",
             transform=None)
    
bbox = ax[3][0].get_tightbbox(fig.canvas.get_renderer())
fig.text(bbox.x0, bbox.y1, 'd.', fontsize=15, fontweight="bold", va="top", ha="left",
             transform=None)


plt.savefig('RBF_supp_configs.pgf', bbox_inches='tight')
plt.savefig('RBF_supp_configs.pdf', bbox_inches='tight')

```
